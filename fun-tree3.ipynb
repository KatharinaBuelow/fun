```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


x=[9.5, 5.5, 13.5, 4.5, 14.5, 3.5, 15.5]
y=[12.5, 8, 8, 5, 5, 2, 2]
# using a point '.' marker
#plt.plot(x,y,linewidth=2, marker ='*', markersize=30, color='yellow')
#plt.plot(x,y, ' ',marker ='*', markersize=30, color='yellow')
```

```python
fig, ax = plt.subplots()

Path = mpath.Path
path_data = [
    (Path.MOVETO, (9., 0.2)),
    (Path.CURVE3, (9., 3)),
    (Path.LINETO, (4., 2.)),
    (Path.CURVE3, (9., 6)),
    (Path.LINETO, (5., 5.)),
    (Path.CURVE3, (9., 9)),
    (Path.LINETO, (6., 8)), 
    (Path.LINETO, (9.5, 12)),
    (Path.LINETO, (13., 8)),
    (Path.CURVE3, (10, 9)),
    (Path.LINETO, (14., 5)),
    (Path.CURVE3, (10, 6)),
    (Path.LINETO, (15., 2)),
    (Path.CURVE3, (10.,3.)),
    (Path.LINETO, (10, 0.2)),
    (Path.CLOSEPOLY, (9., 0.2)),
    ]
codes, verts = zip(*path_data)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='g', alpha=0.5)
ax.add_patch(patch)
plt.plot(x,y, ' ',marker ='*', markeredgecolor='orange', markersize=30, color='yellow')

#ax.grid()

ax.set_xlim(0, 19)
ax.set_ylim(0, 14)
#ax.set_xticks([])
#ax.set_yticks([])
plt.axis('off') 

#ax.axis('equal')
plt.title('Merry Christmas', color='red')
plt.show()
```
